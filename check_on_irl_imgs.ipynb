```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2 # OpenCV is used for color correction

def predict(file_path):
    # --- YOUR FILE PATHS ---
    model_path = r'c:\Users\ABHIRAM\Downloads\best.pt'
    image_path = file_path
    # -----------------------

    # 1. Load your model
    model = YOLO(model_path)

    # 2. Predict on your image
    #    We set show=False because we will plot it manually with Matplotlib
# Raise the IoU threshold to be less strict about overlaps
    results = model.predict(source=image_path, show=False, conf=0.25, iou=0.7)
    # The 'results' object is a list, so we'll work with the first result
    first_result = results[0]

    # 3. Get the result image with bounding boxes
    # The .plot() method returns the image as a NumPy array in BGR format
    image_with_boxes_bgr = first_result.plot()

    # 4. Convert the image from BGR to RGB for correct color display in Matplotlib
    image_with_boxes_rgb = cv2.cvtColor(image_with_boxes_bgr, cv2.COLOR_BGR2RGB)

    # 5. Plot the image
    plt.figure(figsize=(10, 10)) # You can adjust the figure size
    plt.imshow(image_with_boxes_rgb)
    plt.title("Weapon Detection Result")
    plt.axis('off') # Hide the axes for a cleaner look
    plt.show()
```
